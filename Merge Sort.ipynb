```python
def Merge_Sort( arr):
    
    if len(arr) > 1:
        mid = len(arr)//2
        left = arr[:mid]
        right = arr[mid:]
        
        Merge_Sort(left)
        Merge_Sort(right)
        Merge( arr, mid )
```

```python
def Merge( arr, mid):
    
    left = arr[:mid]
    right = arr[mid:]
    
    i=0
    j=0
    k=0
    
    while(i<len(left)  and  j<len(right)):
        if left[i] < right[i]:
            arr[k] = left[i]
            i+=1
        else:
            arr[k] = right[j]
            j+=1
        k+=1
        
    while (i < len(left)):
        arr[k] = arr[i]
        i=i+1
        k=k+1
        
    while (i < len(left)):
        arr[k] = arr[i]
        i=i+1
        k=k+1
```
